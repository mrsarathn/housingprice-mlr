```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# import train_test_split

df = pd.read_csv("https://raw.githubusercontent.com/sonnynomnom/Codecademy-Machine-Learning-Fundamentals/master/StreetEasy/manhattan.csv")


print(df.size)
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.drop(['rental_id', 'rent', 'neighborhood', 'borough' ], axis = 1)
```

```python
y = df.rent
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 6)
```

```python
x_train.shape
```

```python
mlr = LinearRegression()
mlr = mlr.fit(x_train, y_train)

y_predict = mlr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict)
plt.xlabel('Prices: y')
plt.ylabel('Price Predictions')
plt.title('House Prices: Predictions vs Actuals')
plt.plot(y_test, y_test, 'r')
plt.show()
```

```python
print(mlr.coef_)
```

```python
plt.scatter(df.size_sqft, df.rent, alpha = 0.4)
plt.show()
```

```python
sns.pairplot(df, hue = 'rent')
plt.show()
```

```python
print(mlr.score(x_train, y_train))
```

```python
print(mlr.score(x_test, y_test))
```
